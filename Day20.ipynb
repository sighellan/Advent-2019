```python
def lab_f(label_one, label_two):
    if ord(label_one) < ord(label_two):
        return (label_one, label_two)
    return (label_two, label_one)

def find_path_label(yy, xx, grid):
    path_found = False
    path_cor = (-1, -1)
    label_found = False
    label_cor = (-1, -1)
    label = ''
    cor_to_test = [(yy-1, xx), (yy+1, xx), (yy, xx-1), (yy, xx+1)]
    for (cor_yy, cor_xx) in cor_to_test:
        if grid[cor_yy, cor_xx] == '.':
            path_found = True
            path_cor = (cor_yy, cor_xx)
        elif ord(grid[cor_yy, cor_xx]) in range(ord('A'), ord('Z')+1) :
            label_found = True
            label_cor = (cor_yy, cor_xx)
            label = grid[cor_yy, cor_xx]
    return (path_found, path_cor, label_found, label_cor, label)

# Find start location
def find_loc(char_label, grid, hh, ww):
    for yy in range(1, hh-1):
        for xx in range(1, ww-1):
            if grid[yy, xx] == char_label:
                (path_found, path_cor, label_found, label_cor, label) = find_path_label(yy, xx, grid)
                if path_found and label_found and label == char_label:
                    (start_yy, start_xx) = path_cor
                    print((start_xx, start_yy))
                    return((start_yy, start_xx))
```

```python
def next_points_level(yy, xx, level, grid, hh, ww):
    next_pp = []
    cor_to_test_label = [(yy-2, xx), (yy+2, xx), (yy, xx-2), (yy, xx+2)]
    cor_to_test = [(yy-1, xx), (yy+1, xx), (yy, xx-1), (yy, xx+1)]
    for ii in range(len(cor_to_test)):
        (cor_yy, cor_xx) = cor_to_test[ii]
        ll = grid[cor_yy, cor_xx]
        if ll == '.':
            next_pp.append((cor_yy, cor_xx, level))
        elif ord(ll) in range(ord('A'), ord('Z')+1):
            (cor_yy_II, cor_xx_II) = cor_to_test_label[ii]
            ll_II = grid[cor_yy_II, cor_xx_II]
            if ord(ll_II) in range(ord('A'), ord('Z')+1) and (ll_II != ll or (ll not in {'A', 'Z'})):
                tmp_tt = transport_portal_level(yy, xx, level, portal_dict, ll, ll_II, hh, ww)
                if tmp_tt[2] >= 0:
                     next_pp.append(tmp_tt)
            elif ord(ll_II) not in range(ord('A'), ord('Z')+1):
                raise ValueError
    return next_pp

def transport_portal_level(yy, xx, level, portal_dict, lab_one, lab_two, hh, ww):
    label = lab_f(lab_one, lab_two)
    for cc in portal_dict[label]:
        if cc != (yy, xx):
            (new_yy, new_xx) = cc
            old_score = min([yy, hh-yy, xx, ww-xx])
            new_score = min([new_yy, hh-new_yy, new_xx, ww-new_xx])
            if old_score > new_score:
                return (new_yy, new_xx, level+1)
            elif old_score < new_score:
                return (new_yy, new_xx, level-1)
            else:
                print(new_score)
                print(old_score)
                raise ValueError
    
```

```python
def transport_portal(yy, xx, portal_dict, lab_one, lab_two):
    label = lab_f(lab_one, lab_two)
    for cc in portal_dict[label]:
        if cc != (yy, xx):
            return cc
    
    

def next_points(yy, xx, grid):
    next_pp = []
    cor_to_test_label = [(yy-2, xx), (yy+2, xx), (yy, xx-2), (yy, xx+2)]
    cor_to_test = [(yy-1, xx), (yy+1, xx), (yy, xx-1), (yy, xx+1)]
    for (cor_yy, cor_xx) in cor_to_test:
        ll = grid[cor_yy, cor_xx]
        if ll == '.':
            next_pp.append((cor_yy, cor_xx))
        elif ord(ll) in range(ord('A'), ord('Z')+1):
            for (cor_yy_II, cor_xx_II) in cor_to_test_label:
                ll_II = grid[cor_yy_II, cor_xx_II]
                if ord(ll_II) in range(ord('A'), ord('Z')+1) and ll_II != ll:
                    next_pp.append(transport_portal(yy, xx, portal_dict, ll, ll_II))
                    break
    return next_pp
```

```python
# Create data structure of portals
# Dict from labels to paired label and locations

portal_dict = {}

for yy in range(1, hh-1):
    for xx in range(1, ww-1):
        if ord(grid[yy, xx]) in range(ord('A'), ord('Z')+1):
            (path_found, path_cor, label_found, label_cor, label_two) = find_path_label(yy, xx, grid)
            if path_found and label_found:
                label_one = grid[yy, xx]
                label = lab_f(label_one, label_two)
                
                if label in portal_dict.keys():
                    portal_dict[label].append(path_cor)
                else:
                    portal_dict[label] = [path_cor]
            

(start_yy, start_xx) = find_loc('A', grid, hh, ww)
(end_yy, end_xx) = find_loc('Z', grid, hh, ww)
```

```python
# For each start point by a portal, find possible places to walk
options = {}
for pair in portal_dict.values():
#     print(pair)
    for pp in pair:
        ## Find possible steps to take
        pp_yy, pp_xx = pp
        visited = []
        possibilities = []
        queue = [(0, (pp_yy, pp_xx, 5))] # Tuples of next, 
        while len(queue) > 0:
            (dist, (cur_yy, cur_xx, cur_level)) = queue.pop()
            visited.append((cur_yy, cur_xx, cur_level))
            if cur_level != 5:
                if cur_level == 4:
                    lev_diff = -1
                elif cur_level == 6:
                    lev_diff = 1
                else:
                    raise ValueError
                possibilities.append( ((cur_yy, cur_xx), lev_diff, dist) )
            elif (cur_yy, cur_xx) == (end_yy, end_xx):
                possibilities.append( ((cur_yy, cur_xx), 0, dist) )
            else:
                next_pp = next_points_level(cur_yy, cur_xx, cur_level, grid, hh, ww)
                for nn in next_pp:
                    if nn not in visited:
                        queue.append((dist+1, nn))
                        
        options[pp] = possibilities
```
