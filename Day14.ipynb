```python
reac_dict = {}
inp_lines = inp_text.split('\n')
for inp_line in inp_lines:
    ing_list = []
    [ing, result] = inp_line.split(' => ')
    [result_num, result_name] = result.split(' ')
    ingres = ing.split(', ')
    for ii in ingres:
        [num, name] = ii.split(' ')
        ing_list.append((int(num), name))
    reac_dict[result_name] = (int(result_num), ing_list)
```

```python
priority_dict = {'ORE': 1}

while len(priority_dict) < len(reac_dict) + 1:
    for key in reac_dict.keys():
        if key in priority_dict.keys():
            continue
        most_complex = 0
        (_, ingredients) = reac_dict[key]
        all_found = True
        for (_, ing_name) in ingredients:
            if ing_name not in priority_dict:
                all_found = False
                break
            else:
                most_complex = max(most_complex, priority_dict[ing_name])
        if all_found:
            priority_dict[key] = most_complex + 1
    
```

```python
# Parts one and two
start = 1
end = 2000000
# For part 1, set end = 2
# Do a binary search for the number of fuel that can be generated

old_N = 0
while start != end:
    N = (start+end)//2
    if old_N == N:
        break
    cur = {'FUEL': N}
    overflow = {} 
    while list(cur.keys()) != ['ORE']:
        max_jj = list(cur.keys())[0]
        for jj in cur.keys():
            if priority_dict[jj] > priority_dict[max_jj]:
                max_jj = jj
        jj = max_jj
        ing_cur_num = cur.pop(jj)
        reaction = reac_dict[jj]

        if jj in overflow.keys() \
                and overflow[jj] >= min(reaction[0], ing_cur_num):
            to_use = min(overflow[jj], ing_cur_num)
            overflow[jj] = overflow[jj] - to_use
            ing_cur_num = ing_cur_num - to_use


        multiples = np.ceil(ing_cur_num / reaction[0])
        extra = (multiples*reaction[0] - ing_cur_num) % reaction[0]

        if multiples != 0:
            for (ii_num, ii_name) in reaction[1]:
                if ii_name in cur.keys():
                    cur[ii_name] = cur[ii_name] + ii_num*multiples
                else:
                    cur[ii_name] = ii_num*multiples
            if extra != 0:
                if jj in overflow.keys():
                    overflow[jj] = overflow[jj] + extra
                else:
                    overflow[jj] = extra
    print('%s\t%s'%(N, cur['ORE']))
    if cur['ORE'] > 1000000000000:
        end = N
    elif cur['ORE'] < 1000000000000:
        start = N
    else:
        break
    old_N = N
```
